```python
#plotting
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("COVID19.csv")
```

```python
df.head(10)
```

```python
sns.set()
```

```python
miss_count=df.isnull().sum()
print(miss_count)
```

```python
miss_count.plot(kind='bar')
#plot shows that there are a lot of null values..we cant remove all these rows
#lets use imputer class
```

```python
df.info()
```

```python
age_mean=df['age'].mean()
print(age_mean)
```

```python
df['age'].fillna(age_mean,inplace=True)
```

```python
df.info()
```

```python
#now we can perform plotting like heatmap etc which donot support null values on age column
```

```python
#datewise plotting
fig, ax = plt.subplots(figsize=(15,7))

df.groupby(['reporting date','death']).plot();
#sns.relplot(data=df,x="death",y=df.groupby(['reporting date']).count())

```

```python
df2=pd.read_csv("covid_data.csv")

df2_part=df2[['Deaths','SNo','Confirmed','Recovered']]
df1_part=df[['age','id','gender']]
df2_part.head()
```

```python
df1_part.head()
```

```python
df1_part.rename(columns={
        'id': 'SNo', 
        
    }, inplace=True)
data3=pd.merge(df1_part,df2_part)
data3
```

```python
sns.relplot(data=data3,x="Deaths",y="age")
```

```python
sns.relplot(x="Deaths", y="age", size="Deaths",  color="#334488",sizes=(15, 200), data=data3);
```

```python
g = sns.FacetGrid(data3, col="gender", margin_titles=True, height=4)
g.map(plt.scatter, "Deaths", "age", color="#338844", edgecolor="white", s=50, lw=1)
```

```python
#just ignore the graphs with sno
g = sns.PairGrid(data3, hue="gender")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```
